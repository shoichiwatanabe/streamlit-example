# 自分が選んだ画像を加工して楽しめるアプリ

```python
from google.colab import files
uploaded_file = files.upload()
```

```python
uploaded_file_name = next(iter(uploaded_file))
print(uploaded_file_name)
```

```python
import cv2

orig = cv2.imread(uploaded_file_name)
src = cv2.cvtColor(orig, cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt

plt.imshow(src)
```

```python
blurred = cv2.blur(src, (200, 200))
plt.imshow(blurred)
```

```python
import numpy as np

pixels = np.array(src,dtype = 'float32')
pixels = 255 - pixels
plt.imshow(pixels.astype('uint8'))
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.tick_params(labelbottom="off",bottom="off")
ax.tick_params(labelleft="off",left="off")
ax.set_xticklabels([]) 
ax.axis('off')

img = cv.imread('IMG_1479.jpeg', cv.IMREAD_GRAYSCALE)
cv.imwrite('IMG_1479.jpeg', img)
plt.imshow(img)
```
